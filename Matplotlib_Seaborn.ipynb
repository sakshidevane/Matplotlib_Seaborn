# matplotlib basics

```python
import numpy as np    # for arrays   
import pandas as pd     # for dataframe
import matplotlib.pyplot as plt    # for graph visulization
```

```python
x = np.linspace(0,10,10)    # from 0 to 10 gives 10 numbers
x
y = np.linspace(0,20,10)
y
```

```python
x
```

```python
plt.plot(x,y)
plt.title("graph")
plt.xlabel("Numbers")
plt.ylabel("y axis")
```

### subplots 

```python
plt.subplot(2,2,1)        # 2*2 matrix and at position 1 we can take 3*3 or 4*4
plt.plot(x,y)           # 1 2  for subplot position
plt.subplot(2,2,2)      # 3 4
plt.plot(x,y**2)
plt.subplot(2,2,3)
plt.plot(x,x*7)
plt.subplot(2,2,4)
plt.plot(x,x**2)                   
```

### object oriented way

```python
fig= plt.figure(figsize=(20,8),dpi= 100)   # canvas width 20 and height 8
axis1 = fig.add_axes([0.1,0.1,.4,.8])     
# in bracket ([shift from left,shift from bottom, width,hight])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,.2,.3])
axis2.plot(x,y**2)
```

```python
#Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.
#figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

# if dpi= 400 then clearity increases and we can move the graph with axis value
```

# seaborn for quantitative data

## histogram

```python
# We use distribution plots to visualize the distribution of quantitative data.
 # quantitative data = the data which have a value in some range with different values.ex.total bill
  # qualitative data = the data which can be divided into category ex.gender   
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= sns.load_dataset('tips')  # loads the dataset 'tips' from official website of seaborn 
df
```

```python
df['size'].unique()
# as the value are repeting the data can be devided into category so it is qualitative data
```

```python
plt.subplot(1,2,1)   # we can only plot histogram for quantitative data like total_bill and tip
sns.histplot(df['total_bill'],bins= 10, kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'], bins= 50, kde= True)
#bins 
# lables are auto given in seaborn
```

## joint plot

```python
sns.jointplot(x = 'total_bill', y = 'tip', data= df , kind= 'scatter')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data= df , kind= 'reg')
#regression  and kind= scatter, kde,hist,hex,reg,resid
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data= df , kind= 'resid')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data= df , kind= 'kde')
```

## pair plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex') 
# devides the data on basis of sex with color
#sns.pairplot(df,hue='size', palette= 'rainbow') gives size data in different color 
```

# seaborn for categorial data

## countplot

```python
df = sns.load_dataset('tips')
df
```

```python
sns.countplot(df['sex'])  
```

```python
sns.countplot(x=df['sex'], hue = df['smoker'])
# count no of male and female showed on x axis and hue shows male who smoke and not smoke
```

## barplot

```python
sns.barplot(x=df['sex'],  y =df['total_bill'])     # by deafult showing the mean of total bill paid by male and female
```

```python
#You can change the estimator object to your own function, that converts a vector to a scalar Mean (default): np.mean Median: np.median Sum: np.sum Count: len (to count the number of observations) Standard Deviation: np.std Minimum: np.min Maximum: np.max
```

```python
import  numpy as np
sns.barplot(x=df['sex'], y=df['tip'],  estimator= np.sum)
```

## boxplot

```python
sns.boxplot(x= df['tip'],  y=df['day']) 
# small dots are called outliers
```

```python
#boxplots and violinplots are used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.
```

## violinplot

```python
sns.violinplot(x=df['day'], y=df['total_bill'])
# shows sat has more spreaded data 
```

## stripplot and swarmplot

```python
#The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.
#The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).
```

```python
sns.stripplot(x = 'day',y = 'total_bill',data = df)
#sns.tripplot(x=df['day'], y=df['total_bill']) same for above
```

```python
sns.swarmplot(x = 'day',y = 'total_bill',data = df)
```

```python
sns.violinplot(x= 'tip',y = 'day',data = df)
sns.swarmplot(x= 'tip',y = 'day',data = df)
```
